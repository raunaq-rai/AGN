```python
import matplotlib.pyplot as plt
from astropy.io import fits

def plot_spectrum(file_path, title):
    with fits.open(file_path) as hdul:
        data = hdul[1].data
        flux = data['flux']
        loglam = data['loglam']
        wavelength = 10**loglam

        plt.figure(figsize=(10, 6))
        plt.plot(wavelength, flux, label=title)
        plt.xlabel('Wavelength (Angstrom)')
        plt.ylabel('Flux')
        plt.title(title)
        plt.grid(True)
        plt.legend()

# Paths to the original and redshift-corrected spectrum files
original_file = 'spectra1/spec-3586-55181-0920.fits'  
corrected_file = 'redshift_corrected_spectra/spec-3586-55181-0920.fits' 

# Plotting
plot_spectrum(original_file, 'Original Spectrum')
plot_spectrum(corrected_file, 'Redshift-Corrected Spectrum')
plt.show()
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits

def plot_spectra(original_file, corrected_file):
    plt.figure(figsize=(12, 6))

    # Plot the original spectrum
    with fits.open(original_file) as hdul:
        data = hdul[1].data
        flux = data['flux']
        loglam = data['loglam']
        wavelength = 10**loglam
        plt.plot(wavelength, flux, label='Original Spectrum')

    # Plot the redshift-corrected spectrum
    with fits.open(corrected_file) as hdul:
        data = hdul[1].data
        flux = data['flux']
        loglam = data['loglam']
        wavelength = 10**loglam
        plt.plot(wavelength, flux, label='Redshift-Corrected Spectrum')

    plt.xlabel('Wavelength (Angstrom)')
    plt.ylabel('Flux')
    plt.title('Comparison of Original and Redshift-Corrected Spectra')
    plt.axvline(x=4862.721, color='r', linestyle='--', label='4862.721  - Vacuum Wavelength for H-Beta transition')
    plt.grid(True)
    plt.legend()
    plt.savefig('original vs redshift corr.pdf')
    plt.show()

# Paths to the original and redshift-corrected spectrum files
original_file = 'spectra1/spec-3586-55181-0920.fits'  
corrected_file = 'redshift_corrected_spectra/spec-3586-55181-0920.fits'  

# Plotting
plot_spectra(original_file, corrected_file)
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits

def plot_corrected_spectrum(corrected_file):
    plt.figure(figsize=(12, 6))

    # Load and plot the redshift-corrected spectrum
    with fits.open(corrected_file) as hdul:
        data = hdul[1].data
        flux = data['flux']
        loglam = data['loglam']
        wavelength = 10**loglam
        plt.plot(wavelength, flux, label='Redshift-Corrected Spectrum')

    # Define the vacuum wavelengths and labels for the spectral lines
    spectral_lines = {
        'H-beta': 4862.721,
        '[O III] (1)': 4960.295,
        '[O III] (2)': 5008.239,
        '[N II] (1)': 6549.86,
        'H-alpha': 6564.614,
        '[N II] (2)': 6585.27,
        '[S II] (1)': 6718.29,
        '[S II] (2)': 6732.68
    }

    # Add vertical lines for each spectral line
    for line, vacuum_wavelength in spectral_lines.items():
        plt.axvline(x=vacuum_wavelength, color='r', linestyle='--', label=f'{line} {vacuum_wavelength} ')

    # Finalizing the plot
    plt.xlabel('Wavelength (Angstrom)')
    plt.ylabel('Flux')
    plt.title('Redshift-Corrected Spectrum with Spectral Lines')
    plt.grid(True)
    plt.legend()
    plt.savefig('redshift_corr_spectra_with_lines.pdf')
    plt.show()

# Path to the redshift-corrected spectrum file
corrected_file = 'redshift_corrected_spectra/spec-3586-55181-0920.fits' 
plot_corrected_spectrum(corrected_file)
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits

def plot_zoomed_spectrum(corrected_file):
    plt.figure(figsize=(12, 6))

    # Plot the redshift-corrected spectrum
    with fits.open(corrected_file) as hdul:
        data = hdul[1].data
        flux = data['flux']
        loglam = data['loglam']
        wavelength = 10**loglam
        plt.plot(wavelength, flux, label='Redshift-Corrected Spectrum')

    plt.xlabel('Wavelength (Angstrom)')
    plt.ylabel('Flux')
    plt.title('Zoomed Redshift-Corrected Spectrum')
    plt.axvline(x=4862.721, color='r', linestyle='--', label='4862.721  - H-Beta transition')
    plt.axvline(x=4960.295, color='g', linestyle='--', label='4960.295  - O||| (1)')
    plt.axvline(x=5008.239, color='b', linestyle='--', label='5008.239  - O||| (2)')
    plt.xlim(4662.721, 5062.721)
    plt.grid(True)
    plt.legend()
    plt.savefig('zoomed_redshift_corr_spectrum.pdf')
    plt.show()

# Path to the redshift-corrected spectrum file
corrected_file1 = 'redshift_corrected_spectra/spec-3586-55181-0920.fits'

# Plotting the zoomed-in view of the redshift-corrected spectrum
plot_zoomed_spectrum(corrected_file1)
```
