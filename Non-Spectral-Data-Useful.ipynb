```python
from astropy.io import fits
import pandas as pd
```

```python
# Load the FITS file containing the dataset
# 'Rakshit_HbetaSet.fits' should be replaced with the actual file name you are working with
data, header = fits.getdata('Rakshit_HbetaSet.fits', ext=1, header=True)

df = pd.DataFrame(data)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
non_spectral_data = df[['Plate', 'MJD', 'Fiber', 'z', 'logMBH','e_logMBH', 'LogREdd']].copy()
```
